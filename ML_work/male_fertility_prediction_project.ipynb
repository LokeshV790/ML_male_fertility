```python
# Dependencies
import pandas as pd
```

# **ETL: Data Extraction**

```python
# Store filepath in a variable
file = "/content/Data.csv"
```

```python
# Read our Data file with the pandas library
df = pd.read_csv(file)
```

```python
# Show header
df.head()
```

```python
# Change column names
df.columns = ['season', 'age', 'diseases', 'accident', 'surgery', 'fever', 'alcohol', 'smoking', 'sitting', 'diagnosis']
```

```python
df.head()
```

# **ETL: Data Transformation**

```python
#splitting fever column
df_one = pd.get_dummies(df, columns=['fever'], prefix = ['fever'])
df_one.head()
```

```python
#splitting alcohol column
df_two = pd.get_dummies(df_one, columns=['alcohol'], prefix = ['alcohol'])
df_two.head()
```

```python
#splitting smoking column
df_three = pd.get_dummies(df_two, columns=['smoking'], prefix = ['smoking'])
df_three.head()
```

```python
df_three.loc[df_three['sitting'] > 0.5, 'sitting'] = 1
```

```python
df_three.head()
```

```python
df_four = pd.get_dummies(df_three, columns=['sitting'], prefix = ['sitting'])
```

```python
df_five = pd.get_dummies(df_four, columns=['season'], prefix = ['season'])
```

```python
df_five.loc[df_five['age'] > 0.5, 'age'] = 1
```

```python
df_five.loc[df_five['age'] <= 0.5, 'age'] = 0.5
```

```python
df_five.head()
```

```python
df_six = pd.get_dummies(df_five, columns=['age'], prefix = ['age'])
```

```python
df_six.head()
```

```python
df_six.columns
```

```python
#renaming columns:

df_six.rename(columns={
    'fever_-1': 'fever_1yr',
    'fever_0': 'fever_<3mo',
    'fever_1': 'fever_3mo_1yr',
    'alcohol_0.2': 'alcohol_mday',
    'alcohol_0.4': 'alcohol_1day',
    'alcohol_0.6': 'alcohol_mweek',
    'alcohol_0.8': 'alcohol_1week',
    'alcohol_1.0': 'alcohol_negligent',
    'smoking_-1': 'smoke_never',
    'smoking_0': 'smoke_occ',
    'smoking_1': 'smoke_daily',
    'sitting_0.5': 'sitting_<=8hrs',
    'sitting_1.0': 'sitting_>8hrs',
    'season_-1.0': 'season_winter',
    'season_-0.33': 'season_spring',
    'season_0.33': 'season_summer',
    'season_1.0': 'season_fall',
    'age_0.5': 'age_18-27',
    'age_1.0': 'age_28-36'
}, inplace=True)

df_six.columns
```

```python
df_six.drop(columns=['sitting_0.25', 'sitting_0.31', 'sitting_0.38',
       'sitting_0.44', 'sitting_0.47'], inplace=True)
```

```python
df_six.drop(columns=['sitting_0.06', 'sitting_0.13',
       'sitting_0.19'], inplace=True)
```

```python
df_six.head()
```

```python
# Assuming df_six is the dataframe you are working with
df_six = df_six.replace({True: 1, False: 0})

# Verify the transformation
print(df_six.head())
```

```python
df_six.columns
```

```python
X = df_six.drop("diagnosis", axis=1)
y = df_six["diagnosis"]
print(X.shape, y.shape)
```

# **Logistic Regression**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

## **Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
df_one = pd.get_dummies(df, columns=['fever'], prefix = ['fever'])
df_two = pd.get_dummies(df_one, columns=['alcohol'], prefix = ['alcohol'])
df_three = pd.get_dummies(df_two, columns=['smoking'], prefix = ['smoking'])
```

```python
df_three.loc[df_three['sitting'] > 0.5, 'sitting'] = 1
```

```python
df_three.loc[df_three['sitting'] <= 0.5, 'sitting'] = 0.5
```

```python
df_four = pd.get_dummies(df_three, columns=['sitting'], prefix = ['sitting'])
```

```python
df_five = pd.get_dummies(df_four, columns=['season'], prefix = ['season'])
```

```python
df_five.loc[df_five['age'] > 0.5, 'age'] = 1
```

```python
df_five.loc[df_five['age'] <= 0.5, 'age'] = 0.5
```

```python
df_six = pd.get_dummies(df_five, columns=['age'], prefix = ['age'])
df_six.head()
```

```python
# Assuming df_six is the dataframe you are working with
df_six = df_six.replace({True: 1, False: 0})

# Verify the transformation
print(df_six.head())
```

```python
print(df_six.dtypes)
print(df_six.values)
```

```python
X = df_six.drop("diagnosis", axis=1)
y = df_six["diagnosis"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y = label_encoder.transform(y_train)
```

```python
for label, original_class in zip(encoded_y, y):
    print(original_class)
    print(label)
```

```python
from keras.utils import to_categorical
one_hot_y = to_categorical(encoded_y)
one_hot_y
```

```python
model = Sequential()
model.add(Dense(units=6, activation='relu', input_dim=22))
model.add(Dense(units=6, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, one_hot_y, epochs=500, shuffle=True, verbose=2)
```
